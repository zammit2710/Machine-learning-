## Statistical Measures

```python
col=data.Yes
```

```python
#Mean is the average
mean = np.mean(col)
mean
```

```python
# Median is the value of middlemost value
median = np.median(col)
median
```

```python
# Variance is a measure of how variable the data is
variance = np.var(col)
variance
```
